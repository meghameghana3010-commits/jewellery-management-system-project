```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

df = pd.read_csv("/content/home-rental.csv")
print("First 5 rows of the dataset:")
print(df.head())

x=df[["size"]]
y=df[["rent_amount"]]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

model=LinearRegression()
model.fit(x_train,y_train)

y_pred=model.predict(x_test)

mse=mean_squared_error(y_test,y_pred)
print("Mean squared error:",mse)

plt.scatter(x,y,color="skyblue")
plt.plot(x,model.predict(x),color="pink")
plt.xlabel("size")
plt.ylabel("rent_amount")
plt.title("Home Rental Analysis")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

df = pd.read_csv("/content/customers.csv")
print("First 5 rows of the dataset:")
print(df.head())

x=df[["AverageFrequency"]]
y=df[["AverageSpend"]]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

model=LinearRegression()
model.fit(x_train,y_train)

y_pred=model.predict(x_test)

mse=mean_squared_error(y_test,y_pred)
print("Mean squared error:",mse)

plt.scatter(x,y,color="skyblue")
plt.plot(x,model.predict(x),color="red")
plt.xlabel("AverageFrequency")
plt.ylabel("AverageSpend")
plt.title("Customer Spend Analysis")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

df = pd.read_csv("/content/Mall_Customers.csv")

# 2 select numeric columns for clustering
#Replace these columns with numeric columns in your csv
x = df[['Annual Income (k$)','Spending Score (1-100)']]

# 3 create Agglomerative clustering model
agg_model = AgglomerativeClustering(n_clusters=12)

# 4 Fit model and assign cluster labels
df['Cluster'] = agg_model.fit_predict(x)

# 5 Plot clusters
plt.scatter(x['Annual Income (k$)'], x['Spending Score (1-100)'], c=df['Cluster'],cmap='rainbow')
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("Customer Segmentation Example")
plt.show()

df.to_csv("agglo_clustered_data.csv",index=False)
print("Agglomerative clustered dataset saved as 'agglo_clustered_data.csv")
```
